```python
import pandas as pd
from bs4 import BeautifulSoup
import json
import numpy as np
import re
from pandas._libs.missing import NA
from pandas.core.frame import DataFrame
from pandas.tseries.offsets import Second
```

```python
df = pd.read_csv('meta_pages.csv')
```

```python
#define lists to hold data
title = []
description = []
release_date = []
average_critics_rating = []
number_of_critics_reviews = []
user_average_rating = []
number_of_user_reviews = []
content_rating = []
platform = []
other_platforms = []
people_worked_on_game = []
genres = []
publisher = []
number_of_players = []
developer = []
url = []
cover_image = []
```

```python
for index,i in df.iterrows():
    soup = BeautifulSoup(i['page'],'html.parser')
    try:
        title.append(soup.find('div',{'class':'product_title'}).find_next('h1').getText())
    except:
        title.append('')
    #url
    try:
        url.append('https://www.metacritic.com'+soup.find('div',{'class':'product_title'}).find_next('a')['href'])
    except:
        url.append('')
    #cover image
    try:
        cover_image.append(soup.find('img',{'class':'product_image large_image'})['src'])
    except:
        cover_image.append('')
    #description
    try:
        description.append(soup.find('span',{'class':'blurb blurb_expanded'}).getText().replace('\\','').replace('r*',''))
    except:
        description.append('')
    #release_date
    try:
        release_date.append(soup.find('li',{'class':'summary_detail release_data'}).find_next('span',{'class':'data'}).getText())
    except:
        release_date.append('')
    #average_critics_rating
    try:
        average_critics_rating.append(soup.find('span',{'itemprop':'ratingValue'}).getText())
    except:
        average_critics_rating.append('')
    #number_of_critics_reviews
    try:
        number_of_critics_reviews.append(re.sub(r'\W+', '', soup.find('div',{'class':'score_summary metascore_summary'}).find_next('span',{'class':'based'}).find_next('span').getText()).replace('n',''))
    except:
        number_of_critics_reviews.append('')
    #user_average_rating
    try:
        user_average_rating.append(soup.find('div',{'class':'label'},text='User Score').find_next('div').getText())
    except:
        user_average_rating.append('')
    #number_of_user_reviews
    try:
        number_of_user_reviews.append(soup.find('div',{'class':'details side_details'}).find('span',{'class':'based'},text='based on').find_next('a').getText())
    except:
        number_of_user_reviews.append('')
    #content_rating
    try:
        content_rating.append(soup.find('li',{'class':'summary_detail product_rating'}).find_next('span',{'class':'data'}).getText())
    except:
        content_rating.append('')
    #platform
    try:
        platform.append(soup.find('span',{'class':'platform'}).find('a')['href'].replace('/game/',''))
    except:
        platform.append('')
    #other_platforms
    try:
        other_platforms.append(soup.find('li',{'class':'summary_detail product_platforms'}).find_next('a').getText())
    except:
        other_platforms.append('')
    #genres
    try:
        words = ''
        for i in soup.find('li',{'class':'summary_detail product_genre'}).find_all('span',{'class':'data'}):
            words += i.getText() + ','
        genres.append(words[:-1])
    except:
        genres.append('')
    #publisher
    try:
        words = ''

        for i in soup.find('li',{'class':'summary_detail publisher'}).find('span',{'class':'data'}).find_all('a'):
            words += str(i['href']).replace('/company/','').replace('-',' ') + ','
        publisher.append(words[:-1])
    except:
        publisher.append('')
    #number_of_players
    try:
        number_of_players.append(soup.find('li',{'class':'summary_detail product_players'}).find('span',{'class':'data'}).getText())
    except:
        number_of_players.append('')
    #developer
    try:
        developer.append(soup.find('li',{'class':'summary_detail developer'}).find('a').getText())
    except:
        developer.append('')
```

```python
#create temporary dataframe
temp = pd.DataFrame()
#add lists as columns
games = temp.assign(name = title,description = description,release_date = release_date,
average_critics_rating = average_critics_rating,number_of_critics_reviews = number_of_critics_reviews,
user_average_rating = user_average_rating,number_of_user_reviews = number_of_user_reviews,content_rating = content_rating,
platform = platform,other_platforms = other_platforms,
genres = genres,publisher = publisher,number_of_players = number_of_players,developer = developer,url = url,cover_image = cover_image)
```

```python
games.to_csv('meta_cretic.csv',encoding='utf-8',index=False)
```
